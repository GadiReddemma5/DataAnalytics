##PART C: Data Cleaning

```python
retail=retail.dropna(subset='CustomerID')
retail['Description']=retail['Description'].fillna("Unknown Product")
retail.duplicated().sum()
retail.drop_duplicates(inplace=True)
retail=retail[retail['UnitPrice']>0]
retail=retail[retail['Quantity']>0]
retail=retail[~retail['InvoiceNo'].str.startswith('C')]
retail.rename(columns={'Description':"ProductName"},inplace=True)
retail["InvoiceDate"]=pd.to_datetime(retail["InvoiceDate"])
retail.reset_index(inplace=True)
retail
```

##PART D: Feature Creation

```python
retail['TotalPrice']=retail['Quantity']*retail['UnitPrice']
```

```python

retail
```

```python
retail['Year']=retail['InvoiceDate'].dt.year
retail['Month']=retail['InvoiceDate'].dt.month
retail['Day']=retail['InvoiceDate'].dt.day
retail['Hour']=retail['InvoiceDate'].dt.hour
retail['DayofWeek']=retail['InvoiceDate'].dt.day_name()
retail['Quarter']='Q'+retail['InvoiceDate'].dt.quarter.astype(str)
```

```python
conditions=[
    retail["TotalPrice"] < 10,
    (retail["TotalPrice"] >= 10) & (retail["TotalPrice"] <= 100),
    retail["TotalPrice"] > 100
]
choices = ["Low", "Medium", "High"]
retail["RevenueCategory"] = np.select(conditions, choices,default="")
```

```python
conditions=[
    retail["Quantity"] <5,
    (retail["Quantity"] >= 5) & (retail["Quantity"] <=20),
    retail["Quantity"] >20
]
choices = ["Small", "Medium", "Large"]
retail["OrderSize"] = np.select(conditions, choices,default="")
```

##PART H: Aggregation and Summary Statistics

```python
total_revenue = retail["TotalPrice"].sum()
print(total_revenue)
AverageTran=retail["TotalPrice"].mean()
print(AverageTran)
maxTran=retail["TotalPrice"].max()
print(maxTran)
minTran=retail['TotalPrice'].min()

print(minTran)
uniqueCus=retail['CustomerID'].nunique()
print(uniqueCus)
uniquepro=retail['StockCode'].nunique()
print(uniquepro)
uniqueCtry=retail['Country'].nunique()
print(uniqueCtry)
avg_items_per_transaction = retail.groupby("InvoiceNo")["Quantity"].sum().mean()
print(avg_items_per_transaction)
maxReppro=retail['ProductName'].value_counts().idxmax()
print(maxReppro)
```

##PART I: Country-wise Analysis

##PART L: Product Analysis

##PART M: Advanced Grouping

```python
retail.groupby(["Country", "Year"])["TotalPrice"].sum()
retail.groupby(["Country", "InvoiceNo", "Quarter"])["TotalPrice"].sum().groupby(["Country", "Quarter"]).mean()
transactions_country_month = retail.groupby(
    ["Country", "Month"]
)["InvoiceNo"].nunique()
retail.groupby(
    ["DayofWeek", "Hour"]
)["TotalPrice"].sum()
retail.groupby("Country")["TotalPrice"].agg(["sum", "mean", "max", "min"])
retail.groupby(
    ["StockCode"]
).agg(
    total_quantity_sold=("Quantity", "sum"),
    total_revenue=("TotalPrice", "sum")
)
```
